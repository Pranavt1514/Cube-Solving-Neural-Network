```python
import numpy as np

def create_cube_sides():
    # Define colors for each side
    colors = ['w', 'y', 'b', 'g', 'r', 'o']
    
    # Initialize cube sides
    cube_sides = []
    
    # Generate each side randomly
    for color in colors:
        # Create 2D array for the side with all elements as the current color
        side = np.full((3, 3), color)
        
        # Append the side to the cube_sides array
        cube_sides.append(side)
    
    return cube_sides

# Example usage
cube = create_cube_sides()
for side, color in zip(cube, ['white', 'yellow', 'blue', 'green', 'red', 'orange']):
    print(f'{color} side:')
    print(side)
    print()
```

```python
import numpy as np
import random

class RubiksCube:
    def __init__(self):
        self.sides = {
            'w': np.full((3, 3), 'w'),  # White side
            'y': np.full((3, 3), 'y'),  # Yellow side
            'b': np.full((3, 3), 'b'),  # Blue side
            'g': np.full((3, 3), 'g'),  # Green side
            'r': np.full((3, 3), 'r'),  # Red side
            'o': np.full((3, 3), 'o')   # Orange side
        }
        self.solveString= ""
    
    def move(self):
        moves = ['U', 'U\'', 'D', 'D\'', 'L', 'L\'', 'R', 'R\'', 'F', 'F\'', 'B', 'B\'']
        move = random.choice(moves)
        self.solveString+= move
        self._apply_move(move)
    
    def _apply_move(self, move):
        side, direction = move[0], move[1:]
        if direction == '\'':
            direction = -1
        else:
            direction = 1
        #3 axis, 1 white to yellow, 1 blue to green, 1 red to orange, jo axis ke upar side hoga vo 1, nhi to -1
        if side == 'U' or side == 'U\'':
            self._rotate('w', direction)
            self.turn(1, 1, direction)
        elif side == 'D' or side == 'D\'':
            self._rotate('y', direction)
            self.turn(1,-1, direction)
        elif side == 'L' or side == 'L\'':
            self._rotate('b', direction)
            self.turn(0,-1, direction)
        elif side == 'R' or side == 'R\'':
            self._rotate('g', direction)
            self.turn(0,1, direction)
        elif side == 'F' or side == 'F\'':
            self._rotate('r', direction)
            self.turn(-1,1, direction)
        elif side == 'B' or side == 'B\'':
            self._rotate('o', direction)
            self.turn(-1,-1, direction)

    def turn(self, axis, point, direction):
        if axis==1:
            if point==1:
                if direction==1:
                    temp = self.sides['r'][0].copy()
                    self.sides['r'][0]=self.sides['b'][0] 
                    self.sides['b'][0]=self.sides['o'][0]
                    self.sides['o'][0]=self.sides['g'][0]
                    self.sides['g'][0]=temp
                else:
                    temp= self.sides['r'][0].copy()
                    self.sides['r'][0]=self.sides['g'][0] 
                    self.sides['g'][0]=self.sides['o'][0]
                    self.sides['o'][0]=self.sides['b'][0]
                    self.sides['b'][0]=temp
            else:
                if direction==1:
                    temp= self.sides['r'][2].copy()
                    self.sides['r'][2]=self.sides['b'][2] 
                    self.sides['b'][2]=self.sides['o'][2]
                    self.sides['o'][2]=self.sides['g'][2]
                    self.sides['g'][2]=temp
                else:
                    temp= self.sides['r'][2].copy()
                    self.sides['r'][2]=self.sides['g'][2] 
                    self.sides['g'][2]=self.sides['o'][2]
                    self.sides['o'][2]=self.sides['b'][2]
                    self.sides['b'][2]=temp
        if axis==0:
            if point==-1:
                if direction==-1:
                    temp= self.sides['r'][:, 0].copy()
                    self.sides['r'][:,0]=self.sides['y'][:,0] 
                    self.sides['y'][:,0]=self.sides['o'][:,0]
                    self.sides['o'][:,0]=self.sides['w'][:,0]
                    self.sides['w'][:,0]=temp
                else:
                    temp= self.sides['r'][:,0].copy()
                    self.sides['r'][:,0]=self.sides['w'][:,0] 
                    self.sides['w'][:,0]=self.sides['o'][:,0]
                    self.sides['o'][:,0]=self.sides['y'][:,0]
                    self.sides['y'][:,0]=temp
            else:
                if direction==-1:
                    temp= self.sides['r'][:,2].copy()
                    self.sides['r'][:,2]=self.sides['y'][:,2] 
                    self.sides['y'][:,2]=self.sides['o'][:,2]
                    self.sides['o'][:,2]=self.sides['w'][:,2]
                    self.sides['w'][:,2]=temp
                else:
                    temp= self.sides['r'][:,2].copy()
                    self.sides['r'][:,2]=self.sides['w'][:,2] 
                    self.sides['w'][:,2]=self.sides['o'][:,2]
                    self.sides['o'][:,2]=self.sides['y'][:,2]
                    self.sides['y'][:,2]=temp

        if axis==-1:
            if point==1:
                if direction==1:
                    temp= self.sides['y'][0].copy()
                    self.sides['y'][0]=self.sides['b'][0] 
                    self.sides['b'][0]=self.sides['w'][0]
                    self.sides['w'][0]=self.sides['g'][0]
                    self.sides['g'][0]=temp
                else:
                    temp= self.sides['y'][0].copy()
                    self.sides['y'][0]=self.sides['g'][0] 
                    self.sides['g'][0]=self.sides['w'][0]
                    self.sides['w'][0]=self.sides['b'][0]
                    self.sides['b'][0]=temp
            else:
                if direction==1:
                    temp= self.sides['y'][2].copy()
                    self.sides['y'][2]=self.sides['b'][2] 
                    self.sides['b'][2]=self.sides['w'][2]
                    self.sides['w'][2]=self.sides['g'][2]
                    self.sides['g'][2]=temp
                else:
                    temp= self.sides['y'][2].copy()
                    self.sides['y'][2]=self.sides['g'][2] 
                    self.sides['g'][2]=self.sides['w'][2]
                    self.sides['w'][2]=self.sides['b'][2]
                    self.sides['b'][2]=temp
    def _rotate(self, side, direction):
        # Rotate the specified side clockwise (+1) or counterclockwise (-1)
        temp = np.rot90(self.sides[side], direction)
        self.sides[side] = temp
    
    def scramble(self, num_moves=30):
        for _ in range(num_moves):
            self.move()
    
    def print_cube(self):
        for side in ['w', 'y', 'b', 'g', 'r', 'o']:
            print(f'{side} side:')
            print(self.sides[side])
            print()
    def are_similar(self, other):
        for side in self.sides:
            if not np.array_equal(self.sides[side], other.sides[side]):
                return False
      # If all sides are equal, the cubes are similar
        return True
    def print_sol(self):
        return self.solveString

    def flatten_sides(self):
        """
        Flatten the sides arrays into a single string.

        Returns:
        - flattened_string: String containing the flattened sides arrays
        """
        flattened_string = ""
        for key in ['w', 'g', 'b', 'o', 'r', 'y']:
            flattened_string += ' '.join(self.sides[key].flatten()) + ' '
        return flattened_string.strip()

# Create a Rubik's Cube
rubiks_cube = RubiksCube()

# Scramble the cube
rubiks_cube.scramble(1)

# Print the scrambled cube
rubiks_cube.print_cube()
rubiks_cube.print_sol()
```

```python
def solved(cube):
    lst= ['w','o','b','g','y','r']
    for i in range(6):
        block= cube.sides[lst[i]][0][0]
        for j in range(3):
            for k in range(3):
                face= np.full((3, 3), block)
                if not np.array_equal(face, cube.sides[lst[i]]):
                    return False
    return True
solved(rubiks_cube)
```
